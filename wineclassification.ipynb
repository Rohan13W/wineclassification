```python
#importing necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#correlation matrix
corr_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
#define a quality class which indiactes whether is good enough
# 1 if >=7, else 0
df['quality_class'] = df['quality'].apply(lambda x: 1 if x >= 7 else 0)
```

```python
from sklearn.model_selection import train_test_split
X=df.drop(['quality','quality_class'],axis=1)
y=df['quality_class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=26)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

Logistic Regression


```python
from sklearn.linear_model import LogisticRegression
model_logistic = LogisticRegression(solver='lbfgs', max_iter=1000)
model_logistic.fit(X_train, y_train)
```

```python
y_pred_log=model_logistic.predict(X_test)
accuracy_logistic = accuracy_score(y_test, y_pred_log)
print(f'Accuracy: {accuracy_logistic * 100:.2f}%')
print()
conf_matrix = confusion_matrix(y_test, y_pred_log)
print(conf_matrix)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)
```

KNN

```python
y_pred_knn=knn_model.predict(X_test)
accuracy_knn = accuracy_score(y_test, y_pred_knn)
print(f'Accuracy: {accuracy_knn * 100:.2f}%')
print(confusion_matrix(y_test,y_pred_knn))
```

Decision Tree


```python
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)
```

```python
y_pred_dt=dt_model.predict(X_test)
accuracy_dt = accuracy_score(y_test, y_pred_dt)
print(f'Accuracy: {accuracy_dt * 100:.2f}%')
print(confusion_matrix(y_test,y_pred_dt))
```

Random Forest


```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)


```

```python
y_pred_rf=rf_model.predict(X_test)
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f'Accuracy: {accuracy_rf * 100:.2f}%')
print(confusion_matrix(y_test, y_pred_rf))

```

Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
gb_model = GradientBoostingClassifier()
gb_model.fit(X_train, y_train)
```

```python
y_pred_gb=gb_model.predict(X_test)
accuracy_gb = accuracy_score(y_test, y_pred_gb)
print(f'Accuracy: {accuracy_gb * 100:.2f}%')
print(confusion_matrix(y_test,y_pred_gb))
```

```python
from sklearn.model_selection import cross_val_score
k_values = range(1, 45)
cv_scores = []

for k in k_values:
    knn_model = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn_model, X_train, y_train, cv=10, scoring='accuracy')
    cv_scores.append(scores.mean())
plt.plot(k_values, cv_scores)
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Cross-Validation Accuracy')
plt.title('Optimal k for k-NN')
plt.show()
optimal_k = k_values[cv_scores.index(max(cv_scores))]
print(f"The optimal number of neighbors is {optimal_k}")
```

```python
knn_model = KNeighborsClassifier(n_neighbors=optimal_k)
knn_model.fit(X_train, y_train)
```

```python
y_pred_knn=knn_model.predict(X_test)
accuracy_knn = accuracy_score(y_test, y_pred_knn)
print(f'Accuracy: {accuracy_knn * 100:.2f}%')
print(confusion_matrix(y_test,y_pred_knn))
```

```python
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from scipy.stats import randint
param_grid = {
    'n_neighbors': range(1, 51),
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}

param_dist = {
    'n_neighbors': randint(1, 51),
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}

grid_search = GridSearchCV(KNeighborsClassifier(), param_grid, cv=10, scoring='accuracy', verbose=1)
grid_search.fit(X_train, y_train)

random_search = RandomizedSearchCV(KNeighborsClassifier(), param_distributions=param_dist, n_iter=20, cv=10, scoring='accuracy', random_state=42, verbose=1)

random_search.fit(X_train, y_train)
```

```python
print("Grid Search")
print(f"Best Parameters: {grid_search.best_params_}")
print(f"Best Cross-Validation Score: {grid_search.best_score_* 100:.2f}%")

print("RandomSearch")
print(f"Best Parameters: {random_search.best_params_}")
print(f"Best Cross-Validation Score: {random_search.best_score_* 100:.2f}%")
```

```python

best_knn = grid_search.best_estimator_
test_accuracy = best_knn.score(X_test, y_test)
print(f"Test Set Accuracy: {test_accuracy* 100:.2f}%")
y_pred=best_knn.predict(X_test)
print(confusion_matrix(y_test,y_pred))

best_knn_random = random_search.best_estimator_
test_accuracy_random = best_knn_random.score(X_test, y_test)
print(f"Test Set Accuracy: {test_accuracy_random* 100:.2f}%")
y_pred=best_knn_random.predict(X_test)
print(confusion_matrix(y_test,y_pred))
```
